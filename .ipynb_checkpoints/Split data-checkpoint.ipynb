# split the training data

In order to use ImageDataGenerator from Keras each image  belonging to the same  class need to be in one folder. In total we will have 9 folders.

```python
df_train["file_name"] = df_train["img_IDS"] + ".jpg" 
```

```python
 df_train["file_name"]
```

```python
# The '/' will make manipulating files in python difficult, that's why I change the name of this class
df_train["Label"].replace("Enough/Satisfied", "Enough", inplace = True)
```

```python
df_train[df_train["Label"]=="Enough"]
```

Each class will have a subfolder 
You need first to create 9 folders (one for each class) 

```python
TRAIN_FINAL = "C:/Users/HP/Documents/Sign classification/data/final_data" 
```

```python
import os
os.chdir(DESTINATION_FOLDER_TRAIN) 
```

```python
for  file_name in os.listdir(DESTINATION_FOLDER_TRAIN):
    labels =df_train[df_train["file_name"]== file_name]["Label"].values # search for the image's target
    if len(labels)>0: 
        LABEL_FOLDER = os.path.join(TRAIN_FINAL,labels[0])
        shutil.copy(file_name,LABEL_FOLDER)
          
```
